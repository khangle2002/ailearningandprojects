```python
import numpy as np
import matplotlib.pyplot as plt

def numerical_derivative(f, x, h=1e-5):
  f_dev = (f(x+h) - f(x)) / h
  return f_dev

def plot_function_and_derivative(f, x_range=(-2, 2), title=" th f(x) v f(x) ca hm x^2 - 3x + 2"):
    x_vals = np.linspace(x_range[0], x_range[1], 400)
    y_vals = f(x_vals)
    y_derivative_vals = numerical_derivative(f, x_vals)

    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, y_vals, label="f(x)", color="blue")
    plt.plot(x_vals, y_derivative_vals, label="f(x) (numerical)", color="red", linestyle="--")

    # Thm trc ta 
    plt.axhline(0, color="black", linewidth=0.5, linestyle="--")
    plt.axvline(0, color="black", linewidth=0.5, linestyle="--")

    # Ghi ch  th
    plt.xlabel("x")
    plt.ylabel("f(x) & f(x)")
    plt.title(title)
    plt.legend()
    plt.grid()

    # Hin th  th
    plt.show()

def f(x):
    return x**2 - 3*x - 2

plot_function_and_derivative(f, x_range=(-2, 4))
```

```python
def numerical_derivative (f , x , h =1e-5) :
  return ( f ( x + h ) - f ( x ) ) / h

# Hm v  th ca hm s v o hm
def plot_chain_rule (f , g , x_range =( -2 , 2) , title ="o hm Chain Rule ") :
  x_vals = np . linspace ( x_range [0] , x_range [1] , 400)
  # Tnh gi tr ca g(x) v f(g(x))
  g_vals = g ( x_vals )
  f_vals = f ( g_vals )
  # Tnh o hm theo Chain Rule
  f_g_vals = numerical_derivative ( g , x_vals )
  f_f_vals = numerical_derivative ( f , g_vals )
  chain_rule_derivative_vals = f_f_vals * f_g_vals
  # V  th ca f(g(x)) v o hm
  plt . figure ( figsize =(8 , 5) )
  plt . plot ( x_vals , f_vals , label ="f(g(x))", color ="blue")
  plt . plot ( x_vals , chain_rule_derivative_vals , label ="f (x) theo Chain Rule ", color ="red", linestyle ="--")
  plt . axhline (0 , color ="black", linewidth =0.5 , linestyle ="--")
  plt . axvline (0 , color ="black", linewidth =0.5 , linestyle ="--")
  plt . xlabel ("x")
  plt . ylabel ("Gi tr")
  plt . title ( title )
  plt . legend ()
  plt . grid ()

def f(x):
  return math.e**x
def g(x):
  return x**3 - 2*x + 1

plot_chain_rule(f, g, x_range=(-2, 2))
```

```python
import numpy as np

def sigmoid(x):
  return 1/ (1+math.e**(-x))

def sigmoid_derivative(x):
  return (1/ (1+math.e**(-x)))*(1-(1/ (1+math.e**(-x))))

x = np . array ([ -2 , -1 , 0 , 1 , 2])
print (" Sigmoid :", sigmoid ( x ) )

print (" Sigmoid derivative :", sigmoid_derivative ( x ) )
```

```python
def mse_derivative(y_true, y_pred):
    return 2 * (y_pred - y_true) / len(y_true)

y_true = np . array ([3 , 5 , 2 , 8])
y_pred = np . array ([2.5 , 4.8 , 2.1 , 7.5])
print (" Gradient ca MSE :", mse_derivative ( y_true , y_pred ) )
```

```python
def relu(x):
  return np.maximum(0, x)

def relu_derivative(x):
  derivative = []
  for value in x:
    if value > 0:
      derivative.append(1)
    else:
      derivative.append(0)
  return np.array(derivative)

x = np . array ([ -2 , -1 , 0 , 1 , 2])
print (" ReLU :", relu ( x ) )
print (" ReLU derivative :", relu_derivative ( x ) )
```

```python
def tanh(x):
  return (math.e**x - math.e**(-x)) / (math.e**x + math.e**(-x))

def tanh_derivative(x):
  return 1 - tanh(x)**2

x = np . array ([ -2 , -1 , 0 , 1 , 2])
print (" Tanh :", tanh ( x ) )
print (" Tanh derivative :", tanh_derivative ( x ))
```

```python
def softmax(z):
  softmax = []
  total = 0
  for value in z:
    total += math.e**value
  for value in z:
    softmax.append(math.e**value / total)
  return np.array(softmax)

def softmax_derivative(z):
  derisoftmax = []
  total = 0
  for value in z:
    total += math.e**value
  for value in z:
    derisoftmax.append((math.e**value / total)*(1 - math.e**value / total))
  return np.array(derisoftmax)

z = np . array ([2.0 , 1.0 , 0.1])
print (" Softmax :", softmax ( z ) )
print (" Gradient ca Softmax :", softmax_derivative ( z ) )
```
